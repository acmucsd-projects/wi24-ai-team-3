```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

from glob import glob

import librosa
import librosa.display
import IPython.display as ipd
```

```python
audio_files = glob('Dataset/*.wav')
audio_files
```

```python
playaudiofile = ipd.Audio(audio_files[5])
playaudiofile
```

```python
y, sr = librosa.load(audio_files[5])
```

```python
y
```

```python
sr
```

```python
pd.Series(y).plot(figsize=(10, 5),
                  lw = 1,
                  title="audio example")
plt.show()
```

```python
y_trimmed, _ = librosa.effects.trim(y, top_db=60)
```

```python
pd.Series(y_trimmed).plot(figsize=(10, 5),
                  lw = 1,
                  title="trimmed audio example")
plt.show()
```

```python
D = librosa.stft(y)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
img = librosa.display.specshow(S_db,
                               x_axis='time',
                               y_axis='log',
                               ax=ax)
ax.set_title("spectogram ex", fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```

```python
S = librosa.feature.melspectrogram(y=y,
                                   sr=sr,
                                   n_mels=128 * 2,)
S_db_mel = librosa.amplitude_to_db(S, ref=np.max)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
# Plot the mel spectogram
img = librosa.display.specshow(S_db_mel,
                              x_axis='time',
                              y_axis='log',
                              ax=ax)
ax.set_title('Mel Spectogram Example', fontsize=20)
fig.colorbar(img, ax=ax, format=f'%0.2f')
plt.show()
```
